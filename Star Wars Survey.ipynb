## Introduction

While waiting for Star Wars: The Force Awakens to come out, the team at FiveThirtyEight became interested in answering questions about Star Wars fans. In particular: does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?

```python
import pandas as pd
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')

star_wars.head(10)
```

Some of the RespondentID column has null values, since those are unique IDs we would need to eliminate such rows. Some rows also just have "response" labeled instead of an actual response, therefore we have clean those rows as well.

```python
# Review Columns
star_wars.columns
```

Alot of the option for responses to questions are labeled "Unamed: #", which might need adjustment for ease of analysis.

```python
# Remove null ID rows
star_wars[star_wars['RespondentID'].notnull()]
star_wars.head()
```

## Yes/No Questions

The two yes/no questions in the dataset is: 
- Have you seen any of the 6 films in the Star Wars franchise?
- Do you consider yourself to be a fan of the Star Wars film franchise?

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts(dropna=False)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False)
```

From exploring the values in those questions, we conclude that there is a row with a 'Responise' answer which we should delete. We are also going to change the yes/no answer pairs to True/False boolean answer pairs for ease of analysis.

```python
star_wars = star_wars[star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] != 'Response']
star_wars = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] != 'Response']
star_wars.head()
```

```python
yes_no = {'Yes': True, 'No' : False}
cols = ['Have you seen any of the 6 films in the Star Wars franchise?', 'Do you consider yourself to be a fan of the Star Wars film franchise?']

for col in cols:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

## Checkbox Questions

The following are the columns with checkbox answers:
- Which of the following Star Wars films have you seen? Please select all that apply. : Whether or not the respondent saw Star Wars: Episode I The Phantom Menace.
- Unnamed: 4 : Whether or not the respondent saw Star Wars: Episode II Attack of the Clones.
- Unnamed: 5 : Whether or not the respondent saw Star Wars: Episode III Revenge of the Sith.
- Unnamed: 6 : Whether or not the respondent saw Star Wars: Episode IV A New Hope.
- Unnamed: 7 : Whether or not the respondent saw Star Wars: Episode V The Empire Strikes Back.
- Unnamed: 8 : Whether or not the respondent saw Star Wars: Episode VI Return of the Jedi.

We have to rename the columns for the ease of analysis. Also, we are going to perform a similar operation as before to change the answers to True/False. We observe that a response of the movie title name indicates that the repondent did watch the movie, while a null value indicates that the respondent hasn't seen the movie. 

```python
import numpy as np

movie_mapping = {'Star Wars: Episode I  The Phantom Menace':True, 
                 'Star Wars: Episode II  Attack of the Clones':True,
                 'Star Wars: Episode III  Revenge of the Sith':True,
                 'Star Wars: Episode IV  A New Hope':True,
                 'Star Wars: Episode V The Empire Strikes Back':True,
                 'Star Wars: Episode VI Return of the Jedi':True,
                  np.NaN: False}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)

star_wars.head()
```

Rename the columns to make it more intuitive. Use seen_1 to indicate that the respondent has seen Star Wars: Episode I The Phantom Menace, seen_2 to indicate that the respondent has seen Star Wars: Episode II Attack of the Clones, and so on.

```python
rename_cols = {
    'Which of the following Star Wars films have you seen? Please select all that apply.':'seen_1',
    'Unnamed: 4':'seen_2',
    'Unnamed: 5':'seen_3',
    'Unnamed: 6':'seen_4',
    'Unnamed: 7':'seen_5',
    'Unnamed: 8':'seen_6'}

star_wars = star_wars.rename(columns=rename_cols)
star_wars.head()
```

## Ranking Columns

The following are columns with ranking answers:
- Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film. : How much the respondent liked Star Wars: Episode I The Phantom Menace
- Unnamed: 10 : How much the respondent liked Star Wars: Episode II Attack of the Clones
- Unnamed: 11 : How much the respondent liked Star Wars: Episode III Revenge of the Sith
- Unnamed: 12 : How much the respondent liked Star Wars: Episode IV A New Hope
- Unnamed: 13 : How much the respondent liked Star Wars: Episode V The Empire Strikes Back
- Unnamed: 14 : How much the respondent liked Star Wars: Episode VI Return of the Jedi

The ranking columns ask the respondent to rank the movies. 1 being their favorite and 6 being their least favorite. Those columns are relatively clean, therefore we just need to change them to numeric type data and rename the columns. The columns are going to be renamed to ranking_1 for the ranking on movie 1 and ranking_2 for the ranking on movie 2 etc.

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

rename_columns = {
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ranking_1',
    'Unnamed:10':'ranking_2',
    'Unnamed:11':'ranking_3',
    'Unnamed:12':'ranking_4',
    'Unnamed:13':'ranking_5',
    'Unnamed:14':'ranking_6'
}

star_wars = star_wars.rename(columns=rename_columns)
star_wars.head()
```

## Ranking Analysis

```python
# Calculate mean for ranking columns
star_wars[star_wars.columns[9:15]].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.bar(range(1,7), star_wars[star_wars.columns[9:15]].mean())
```

Since a lower ranking indicates that the movie is ranked higher, we see that Star Wars: Episode V The Empire Strikes Back is clearly ranked the highest of all movies. Episodes 4 and 6 also has fairly high rankings.

## Seen Columns Analysis

Since we changed all the "seen" columns to boolean values, we are able to compute the amount of people who saw each movie by summing the rows. 

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(1,7),star_wars[star_wars.columns[3:9]].sum())
```

The most seen movies are Episodes 5 and 6. There might be a positive correlation between the movie ranking and the amount of people who has seen the movie.

## Binary Audience Segments

We have analyzed how the movies are ranked by the survey population. The dataset also includes more segmented audiences: 
- Do you consider yourself to be a fan of the Star Wars film franchise? - True or False
- Gender - Male or Female

We can do a further analysis by seperating the audience into groups. 

```python
# Gender Segments
males = star_wars[star_wars['Gender']=='Male']
females= star_wars[star_wars['Gender']=='Female']
```

```python
# Plot the most seen movies by gender
fig = plt.figure(figsize=(13,4))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.bar(range(1,7),males[males.columns[3:9]].sum())
ax1.set_title('males')
ax2.bar(range(1,7),females[females.columns[3:9]].sum())
ax2.set_title('females')
```

```python
# Plot the rankings of the movie by gender
fig = plt.figure(figsize=(13,4))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.bar(range(1,7),males[males.columns[9:15]].mean())
ax1.set_title('males')
ax2.bar(range(1,7),females[females.columns[9:15]].mean())
ax2.set_title('females')
```

On average, males watched episodes 1-3 more than females, and the rankings of episodes 4-6 are higher in both genders. 

```python
# Fan segmentations
fan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']==True]
non_fan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']==False]
```

```python
# Plot the most seen movie by fan 
fig = plt.figure(figsize=(13,4))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.bar(range(1,7),fan[fan.columns[3:9]].sum())
ax1.set_title('fans')
ax2.bar(range(1,7),non_fan[non_fan.columns[3:9]].sum())
ax2.set_title('non-fans')
```

```python
# Plot the movie rankings by fan
fig = plt.figure(figsize=(13,4))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.bar(range(1,7),fan[fan.columns[9:15]].mean())
ax1.set_title('fans')
ax2.bar(range(1,7),non_fan[non_fan.columns[9:15]].mean())
ax2.set_title('non-fans')
```

Fans watch more of the movies on average, which is intuitive. Fans are more likely to rank episodes 4-6 higher, which indicates their preference for the movies from the franchise.

## Non-Binary Audience Segments

Next we want to perform some analysis on segments that are non-binary and see if there are interesting results. Specifically, we are going to focus on the 'Education' column.

```python
# Explore education column
star_wars['Education'].value_counts()
```

We could see that the main audience of the star wars movies has some college experience. Since audience group that has less than a high school degree is so low, we are going to disregard it during our analysis.

```python
some_college = star_wars[star_wars['Education']=='Some college or Associate degree']
bachelor = star_wars[star_wars['Education']=='Bachelor degree']
graduate = star_wars[star_wars['Education']=='Graduate degree']
high_school = star_wars[star_wars['Education']=='High school degree']
```

```python
# Plot the amount of movie watched based on education
fig = plt.figure(figsize=(15,13))

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.bar(range(1,7),some_college[some_college.columns[3:9]].sum())
ax2.bar(range(1,7),bachelor[bachelor.columns[3:9]].sum())
ax3.bar(range(1,7),graduate[graduate.columns[3:9]].sum())
ax4.bar(range(1,7),high_school[high_school.columns[3:9]].sum())
```

Proportionally speaking, we see that episodes 5 and 6 are more watched on average than the ealier episodes which aligns with what we observed in previous analysis on the whole survey group.
